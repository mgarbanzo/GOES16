# Lectura de datos


```python
from netCDF4 import Dataset
import pylab as plt
```

```python
archivo = Dataset("/content/OR_GLM-L2-LCFA_G16_s20211892330000_e20211892330200_c20211892330219.nc")
```

```python
latitudes = archivo.variables['event_lat'][:]
longitudes = archivo.variables['event_lon'][:]
```

```python
plt.plot(latitudes,'ok')
```

```python
plt.plot(longitudes,'ok')
```

```python
HistLats = plt.hist(latitudes, bins=100)
```

```python
HistLons = plt.hist(longitudes, bins=100)
```

```python
plt.scatter(longitudes,latitudes)
```

# Agregado energía

```python
energia = archivo.variables['event_energy'][:]
```

```python
plt.plot(energia,'ok')
```

```python
HistEnergia = plt.hist(energia, bins=100)
```

```python
factorEscala = 1E15
plt.scatter(longitudes,latitudes,s=energia*factorEscala,marker='+')
```
